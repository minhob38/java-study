```python
"""
자료형이란 데이터를 표현하는 기준을 의미하며, 파이썬에는 숫자형 / 문자열 / 리스트 / 튜플 / 딕셔너리 / 집합 / 불이 있습니다.
"""
print("\n=== 자료형 ===")
```

```python
"""
숫자형
(숫자형에는 정수/실수/8진수/16진수가 있습니다.)
"""
# 정수형 
print("\n=== 정수형 ===")
n = 3
print(n, type(n))

# 8진수(정수형)
n = 0o37
print(n, type(n))

# 16진수(정수형)
n = 0x3f
print(n, type(n))

# 실수형 
print("\n=== 실수형 ===")
n = 3.5
print(n, type(n))

n = 3e5
print(n, type(n))
```

```python
"""
문자열
(문자열은 " ", ' ', """ """, ''' '''로 만듭니다.)
"""
print("\n=== 문자열 만들기 ===")
s = "python"
print(s, type(s))

s = 'python'
print(s, type(s))

s = """python"""
print(s, type(s))

s = '''python'''
print(s, type(s))

# 문자열 가져오기
print("\n=== 문자열 가져오기. ===")
s = "python"
print(s[1])
print(s[2:5])
print(s[2:])
print(s[:5])
print(s[-1])
print(s[:-1])
print(s[-6:])
print(s[-6:-1])
print(s[::-1]) # 문자열 뒤집기

# 문자열은 immutable 자료형으로, 가져온 문자에 새로운 문자를 할당할 수 없습니다.
print("\n=== 문자열은 immutable 자료형으로, 가져온 문자에 새로운 문자를 할당할 수 없습니다. ===")
try:
    s = "python"
    s[1] = "m"
except Exception as err:
    print(err)
    
# 문자열 포맷팅
print("\n=== 문자열 포맷팅 ===")
print("python %o" %0o27)
print("python %x" %0x2E)
print("python %d %f %s" %(1, 1.1, ": )"))

print("python {}".format(": )"))
print("python {} {}".format(": )", ": ("))
print("python {1} {0}".format(": )", ": ("))
print("python {p} {q}".format(p=": )", q=": ("))

s = ": )"
print(f"python {s}")

# 이스케이프 문자
print("\n=== 이스케이프 문자 ===")
print("py\"thon")
print("py\nthon")
```

```python
"""
리스트
(리스트란 공통 특성을 가진 자료형들을 모아 넣은 자료형입니다.)
"""
print("\n=== 리스트 만들기 ===")
l = [1, 2, 3]
print(l, type(l))

l = ["a", "b", "c"]
print(l, type(l))

l = ["a", 1, True]
print(l, type(l))

l = ["a", 1, [True, None]]
print(l, type(l))

# 리스트 가져오기
print("\n=== 리스트 가져오기. ===")
l = ["a", 1, True, None, "3", [5, "b"]]
print(l[1])
print(l[2:5])
print(l[2:])
print(l[:5])
print(l[-1])
print(l[:-1])
print(l[-6:])
print(l[-6:-1])
print(l[::-1]) # 리스트 뒤집기

# 리스트 수정하기
print("\n=== 리스트 수정하기 ===")
l = ["a", 1, True, None, "3", [5, "b"]]

l[1] = 10
print(l, l[1])

l[5][0] = 15
print(l, l[5][0])

# 리스트 지우기
print("\n=== 리스트 지우기 ===")
l = ["a", 1, True, None, "3", [5, "b"]]

del l[1]
print(l, l[1])

del l[4][0]
print(l, l[4][0])

del l[1:3]
print(l)
```

```python
"""
딕셔너리
(딕셔너리란 key-value를 담을 수 있는 자료형입니다.)
"""
# 딕셔너리 만들기
print("\n=== 딕셔너리 만들기 ===")
d = { "a": 1, "b": "python", 3: "django", "3": True, "5": "flask" }
print(d, type(d))

# 딕셔너리 가져오기
print("\n=== 딕셔너리 가져오기 ===")
d = { "a": 1, "b": "python", 3: "django", "3": True, "5": "flask" }

print(d["a"], d["b"], d[3], d["3"], d["5"])

# 딕셔너리 지우기
print("\n=== 딕셔너리 지우기 ===")
d = { "a": 1, "b": "python", 3: "django", "3": True, "5": "flask" }

del d["a"]
del d[3]
print(d)

# 딕셔너리 조회
print("\n=== 딕셔너리 조회하기 ===")
print(d.keys(), type(d.keys())) # value들을 list로 반환
print(d.values(), type(d.values())) # value들을 list로 반환
print(d.items(), type(d.items())) # key-value들의 tuple을 list로 반환
```

```python
"""
튜플
(튜플은 리스트와 비슷하지만, immutable한 차이가 있습니다.)
"""
# 튜플 만들기
print("\n=== 튜플 만들기 ===")
t = (5, 8)
print(t, type(t))

t = (5, "python", True)
print(t, type(t))

# 튜플은 요소가 두개 이상이어야 합니다.
print("\n=== 튜플은 요소가 두개 이상이어야 합니다. ===")
t = (5)
print(t, type(t))

t = (5,)
print(t, type(t))


# 튜플은 immutable 자료형입니다.
print("\n=== 튜플은 immutable 자료형입니다. ===")
try:
    t = (5, "python", True)
    t[1] = "m"
except Exception as err:
    print(err)
```

```python
"""
set
(set은 중복이 없는 자료형입니다.)
"""
# 튜플 만들기
print("\n=== set 만들기 ===")
s = set()
print(s, type(s))

s.add(5)
print(s, type(s))

s.add("python")
print(s, type(s))

s.add("python")
print(s, type(s))

s = set([5, "python", "python"])
print(s, type(s))
```

```python
"""
불
(참과 거짓을 표현하는 자료형으로 참은 True, 거짓은 False로 표현합니다.)
"""
# True / False
print("\n=== True / False ===")
print(True)
print(False)
print(3 == 3)
print(3 != 3)

# Truthy / Falsy
print("\n=== Truthy / Falsy ===")
# Truthy
print("\n=== Truthy ===")
print(bool("python"))
print(bool(1)) 
print(bool(-1))
print(bool([1]))
print(bool({ "a": 3 }))
print(bool((0,)))

# Falsy
print("\n=== Falsy ===")
print(bool(""))
print(bool(0))
print(bool(0.0))
print(bool([]))
print(bool({}))
print(bool(()))
print(bool(None))
```
