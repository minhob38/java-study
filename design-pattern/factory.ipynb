```python
from abc import *

class Shoes():
    @abstractmethod
    def sew(self):
        pass
    
    @abstractmethod
    def pack(self):
        pass
    
class FootballShoes(Shoes):
    def sew(self):
        print("football sewing")
     
    def pack(self):
        print("football packing")       

        
class BasketballShoes(Shoes):
    def sew(self):
        print("basketball sewing")
        
    def pack(self):
        print("basketball packing")
```

```python
"""
=== simple factory ===
instance(객체)를 만드는 class를 factory class라 하며, 이때 단순히 instance를 만드는 factory를 simple factory라 합니다.
- simple factory로 단순히 instance만 만들면, instance를 어떻게 다루는지에 대해 모자란 문제가 있습니다.
- 예로들어 shoes instance를 simple factory로 만들어도, 이를 소비하는 class에서 sew, pack method를 실행하지 않아 shoes instance가 이상한 형태가 될수 있습니다.)
"""

# shoes factory는 shoes instance를 만듭니다.
class ShoesSimpleFactory():        
    def create_shoes(self, type) -> Shoes:
        if (type == "football"):
            return FootballShoes()
        if (type == "basketball"):
            return BasketballShoes()

# shoes store는 주문(oder)이 들어오면 신발을 바느질(sewing), 포장(packing)하여 신발을 준비합니다.
class ShoesStore():
    def __init__(self, shoes_simple_factory):
        self.shoes_simple_factory = shoes_simple_factory

    def order(self, type):
        shoes = self.shoes_simple_factory.create_shoes(type)
        shoes.sew()
        shoes.pack()

shoes_simple_factory = ShoesSimpleFactory()
shoes_store = ShoesStore(shoes_simple_factory)
shoes_store.order("football")
shoes_store.order("basketball")

class NikeShoesStore(ShoesStore):
    def __init__(self, shoes_simple_factory):
        self.shoes_simple_factory = shoes_simple_factory

    def order(self, type):
        shoes = self.shoes_simple_factory.create_shoes(type)
        shoes.sew()
    #   shoes.pack() - 포장을 안함
    
class AdidasShoesStore(ShoesStore):
    def __init__(self, shoes_simple_factory):
        self.shoes_simple_factory = shoes_simple_factory

    def order(self, type):
        shoes = self.shoes_simple_factory.create_shoes(type)
    #   shoes.sew() - 바느질을 안함
        shoes.pack()
        
nike_shoes_store = NikeShoesStore(shoes_simple_factory)
nike_shoes_store.order("football")

adidas_shoes_store = AdidasShoesStore(shoes_simple_factory)
adidas_shoes_store.order("football")
```

```python
"""
=== factory method ===
superclass에서는 instance interface만 정의하고, subclass에서 instance를 만들도록 합니다.
- factory(store)에서 단순히 instance 생성뿐만 아니라, instance를 어떻게 다루는지까지 형식화해줍니다.
- factory(store) subclass에서 어떤 instance를 만들지 정의만 해주면됩니다.
"""

class ShoesStore(): # <- factory class
    @abstractmethod
    def create_shoes(self) -> Shoes:
        pass
    
    def order(self, type):
        shoes = self.create_shoes(type)
        shoes.sew()
        shoes.pack()

class NikeShoesStore(ShoesStore):
    def create_shoes(self, type):
        if (type == "football"):
            return FootballShoes()
        if (type == "basketball"):
            return BasketballShoes()

class AdidasShoesStore(ShoesStore):
    def create_shoes(self, type):
        if (type == "football"):
            return FootballShoes()
        if (type == "basketball"):
            return BasketballShoes()
        
nike_shoes_store = NikeShoesStore()
nike_shoes_store.order("football")
nike_shoes_store.order("basketball")

adidas_shoes_store = AdidasShoesStore()
adidas_shoes_store.order("football")
adidas_shoes_store.order("basketball")
```
