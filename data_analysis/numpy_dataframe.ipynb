```python
# numpy array calculation
na = np.arange(-np.pi, np.pi, np.pi / 100)
plt.plot(na, np.sin(na))
plt.plot(na, np.cos(na))
plt.plot(na, np.cos(na - np.pi / 4))
plt.show()
```

```python
# numpy mask
na = np.arange(-10, 10)
mask = na > 0
print(mask)
print(na[mask])

mask_a = abs(na) > 3
mask_b = abs(na) % 2 == 0
print(na[mask_a + mask_b])
print(na[mask_a * mask_b])
```

```python
# numpy mask plot
x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)
size = np.random.rand(1000) * 100

mask_a = abs(x) > 50
mask_b = abs(y) > 50
x = x[mask_a + mask_b]
y = y[mask_a + mask_b]
size = size[mask_a + mask_b]

plt.scatter(x, y, s=size, c=x, cmap="jet", alpha=0.5)
plt.colorbar()
plt.show()
```

```python
# 해당 시군구와 유사한 시군구의 연령별 인구 구조 비교 plot
f = open("./data/sgg_population.csv", "r", encoding="utf-8")
data = csv.reader(f, delimiter=",")

header = next(data)

sgg = input("인구를 알고 싶은 시군구를 입력해주세요 : )") #e.g 수지구, 기흥구
min_diff = 1
away = np.zeros(0)
away_name = ""

for row in data:
    if sgg in row[0]:
        home = np.array(list(map(lambda item: item.replace(",", ""), row[3:104])), dtype=int) / int(row[2].replace(",", ""))
        break
        
for row in data:
    row[2:104] = list(map(lambda item: int(item.replace(",", "")) if "," in item else int(item), row[2:104] ))

    if row[2] != 0:
        comp = np.array(row[3:104]) / row[2]
        diff = np.sum((home - comp)**2)

        if diff < min_diff and sgg not in row[0]:
            min_diff = diff
            away = comp
            away_name = row[0].split()[-2]
    
plt.rc('font', family='AppleGothic')
plt.title(f"{sgg} 인구구조와 비슷한 시군구")
plt.plot(home, label=sgg)
plt.plot(away, label=away_name)
plt.legend(loc=1)
plt.show()

f.close()
```
