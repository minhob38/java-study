```python
# 해당 시군구의 연령별 인구 구조 plot
df = pd.read_csv("./data/sgg_population.csv", encoding="utf8", index_col=0, thousands=",")
df = df.iloc[:,:103]
df = df.div(df["2021년01월_계_총인구수"], axis=0)
sgg = input("인구를 알고 싶은 시군구를 입력해주세요 : )") #e.g 수지구, 기흥구
mask = df.index.str.contains(sgg)
del df["2021년01월_계_총인구수"]
del df["2021년01월_계_연령구간인구수"]

df_sgg = df[mask]

plt.rc('font', family='AppleGothic') 
df_sgg.T.plot() # index: x-axis / column y-axis
plt.xticks(rotation=90)
plt.title(f"{sgg}'s population")
plt.show()
```

```python
# 해당 시군구의 연령별 인구 구조 plot
"""
dataframe 연산에는 series로 해야함
print(type(df_sgg))
print(type(df_sgg.iloc[0]))
"""
df = pd.read_csv("./data/sgg_population.csv", encoding="utf8", index_col=0, thousands=",")
df = df.iloc[:,:103]
df = df.div(df["2021년01월_계_총인구수"], axis=0)
# df = df.fillna(99999)
df = df.dropna()
sgg = input("인구를 알고 싶은 시군구를 입력해주세요 : )") #e.g 수지구, 기흥구
mask = df.index.str.contains(sgg)
del df["2021년01월_계_총인구수"]
del df["2021년01월_계_연령구간인구수"]

df_sgg = df[mask]

p = df.sub(df_sgg.iloc[0], axis=1)
q = np.power(p, 2)
r = np.sum(q, axis=1)
away = r.sort_values().index[:5]

plt.rc('font', family='AppleGothic') 
df.loc[away].T.plot()
plt.xticks(rotation=90)
plt.title(f"{sgg}'s population")
plt.show()
```
